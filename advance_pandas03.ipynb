```python
pints_df=pd.DataFrame({
    '1st point':[(1,2),(4,5),(5,4),(7,9),(2,3)],
    '2nd point':[(3,5),(4,4),(6,5),(8,9),(7,7)]
})
```

```python
pints_df
```

```python
def euclidean(row):
  pt_a=row['1st point']
  pt_b=row['2nd point']
  return ((pt_a[0]-pt_b[0])**2)+((pt_a[1]-pt_b[1])**2)**0.5
```

```python
pints_df['distance']=pints_df.apply(euclidean,axis=1)
```
