#3d plotting

```python
#3d scatter plot
fig=plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(batter['avg'],batter['strike_rate'],batter['runs'])
```

```python
#LINE PLOT
x=[0,1,5]
y=[0,10,13]
z=[0,13,20]

fig=plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot3D(x,y,z)
```

```python
#3D SURFACE PLOT
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
Z=xx**2+yy**2
fig=plt.figure()

ax=plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
z=np.sign(xx)+np.cos(yy)
ax=plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
#contour plot
fig=plt.figure()
ax=plt.subplot()
p=ax.contourf(xx,yy,z,50)
fig.colorbar(p)
```

```python
delivery=pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
```
