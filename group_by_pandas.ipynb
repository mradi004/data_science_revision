```python
#looping on groups
df=pd.DataFrame(columns=imdb.columns)
for group,data in genres:
  # Use pd.concat instead of append
  df=pd.concat([df, data[data['IMDB_Rating']==data['IMDB_Rating'].max()]])
```

```python
#apply function
genres.apply(min)
```

```python
#find the movies starting with A for each group
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
#find ranking of each movie in the group according to imdb rating
def rank(group):
  group['ranking']=group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank)
```

```python
#noramlised rating
def too(group):
  group['normalise_rating']=(group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
  return group
```

```python
genres.apply(too)
```

```python
def high(batsman):
  tempdff=ipl[ipl['batsman']==batsman]
  return tempdff.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```
