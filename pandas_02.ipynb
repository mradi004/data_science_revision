#Apply function in Pandas

```python
#we are going to use apply function
adi['Fare_INR']=adi['Fare'].apply(lambda x:x*82)
```

```python
adi
```

```python
adi.head()
```

```python
adi['len_name']=adi['Name'].apply(lambda x:len(x))
```

```python
adi['len_name'].head()
```

```python
def cat_fare(x):
  if x<10:
    return 'cheap'
  elif x>=10 and x<20:
    return 'mid'

  else:
    return 'high'
```

```python
adi['cat_fare']=adi['Fare'].apply(cat_fare)
```
