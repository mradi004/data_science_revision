```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,200)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title('this is my title')
plt.xlabel('this is my x data')
plt.ylabel('this is my y data')
plt.xlabel("this is my new x data")
plt.ylabel("this is my new y data")
plt.show()
```

```python
x=np.random.rand(50)

y=np.random.rand(50)
```

```python
plt.scatter(x,y)
```

```python
x
```

```python
colurs=np.random.rand(50)
sizes=np.random.rand(50)*1000
plt.scatter(x,y,c=colurs,s=sizes,alpha=0.9)
plt.xlabel("this is my new x data")
plt.ylabel("this is my new y data")
plt.show()
```

```python
x=['a','b','c','d','e']
y=[1,5,6,7,8]
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
x=np.random.rand(50)
y=np.sin(x)
```

```python
plt.figure(figsize=(4,5))
plt.plot(x,y,'--b')
plt.xlabel("this is my x")
plt.ylabel('this is my y')
plt.show()
```

```python
x=[4,5,6,7,8,9]
y=[45,65,34,54,65,45]
plt.plot(x,y)
plt.show()
```

```python
colurs=['red','blue','yellow','pink','black','olive']
plt.scatter(x,y,c=colurs)
plt.show()
```

```python
#plotting of 3d data
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
ax.show()
```
