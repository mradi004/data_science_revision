```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.1)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.1)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.1)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C']
men = [20, 35, 30]
women = [25, 32, 34]

# Plot
plt.bar(categories, men, label='Men')
plt.bar(categories, women, bottom=men, label='Women')  # stacked

plt.ylabel('Scores')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']),label='2015')
plt.xticks(rotation=90)

plt.legend()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]

# Plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.axis('equal')  # Makes the pie round
plt.show()
```
