```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#scatter plot
titanic.dropna(subset=['age'], inplace=True)
fig=px.scatter(titanic,x='age',y='fare',color='sex',size='fare')
fig.show()
fig.write_image('scatter_plot.jpeg',scale=3,width=1800,height=900)
```

```python
import plotly.graph_objects as go
```

```python
fig=go.Figure(data=go.Scatter(x=titanic['age'],y=titanic['fare'],mode='lines'))
fig.show()
```

```python
#heatmaps
data=np.random.rand(10,10)
data
fig=go.Figure(data=go.Heatmap(z=data))
fig.show()
```

```python
#heatmap on correlation of age and fareusing plotly
fig=px.imshow(titanic[['age','fare']].corr())
fig.show()
```

```python
#scatter plot
fig1=px.scatter(iris,x='sepal_width',y='sepal_length',color='species')
fig1.show()
fig1.write_image('scatter_plot.jpeg',scale=3,width=1800,height=900)
```

```python
#line plot
sortediris=iris.sort_values(by='sepal_width')
fig2=px.line(sortediris,x='sepal_width',y='sepal_length',color='species')
fig2.show()
fig2.write_image('line_plot.jpeg',scale=3,width=1800,height=900)
```

```python
#bar plot
#grup values based on means
meanediris=iris.groupby('species').mean().reset_index()
fig3=px.bar(meanediris,x='species',y='sepal_width',color='species')
fig3.show()
```

```python
#box plot
fig4=px.box(iris,x='species',y='sepal_width',color='species')
fig4.show()
```

```python
#violin plot
fig5=px.violin(iris,x='species',y='sepal_width',color='species')
fig5.show()
```

```python
#histogram
fig6=px.histogram(iris,x='species',color='species')
fig6.show()
```

```python
#pie chart
fig7=px.pie(iris,values='sepal_width',names='species')
fig7.show()
```

```python
#scatter 3d
fig8=px.scatter_3d(iris,x='sepal_width',y='sepal_length',z='petal_width',color='species')

fig8.show()
```

```python
#area chart
```

```python
#bubble charft
fig10=px.scatter(iris,x='sepal_width',y='sepal_length',color='species',size='petal_width')
fig10.show()
```

```python
#sunburst chart
df_sunburst=iris.groupby(['species','petal_width']).mean().reset_index()
fig11=px.sunburst(df_sunburst,path=['species','petal_width'],values='species_id')
fig11.show()
```
