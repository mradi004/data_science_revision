```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

#Relational Plot

```python
#figure and axes functions
tips=sns.load_dataset("tips")
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue='sex',style='time',size='size')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python
#line plot
gap=px.data.gapminder()
temp=gap[gap['country']=='India']
sns.lineplot(data=temp,x='year',y='lifeExp')
```

```python
sns.relplot(data=temp,x='year',y='lifeExp',kind='line')
```

```python
temp_df=gap[gap['country'].isin(['India','Pakistan','China'])]
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country')
```

```python
#facet plot
sns.relplot(kind='scatter',data=tips,x='total_bill',y='tip',col='sex',hue='time')
```

```python
sns.relplot(kind='scatter',data=gap,x='gdpPercap',y='lifeExp',col='continent',hue='country')
```

#multiplots


```python
#facetgrid
g=sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
```

```python
#pairplot and pairgrids
sns.pairplot(data=tips,hue='day')
```

```python
sns.pairplot(data=iris,hue='species')
```

```python
#pairgrid
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
#jointgrid and jointplot
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
g=sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.violinplot)
```
