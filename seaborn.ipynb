```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns
import warnings
```

```python
iris=sns.load_dataset('iris')
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.petal_length)
```

```python
sns.displot(iris.sepal_length)
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="size",style="smoker",data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,col='time',kind='bar',height=2,aspect=2)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,height=2)
```

```python
sns.pairplot(tips)
```
